```python
def delete_column_by_name(file_path, column_name):
    rows = []
    with open(file_path, 'r', newline='') as csvfile:
        reader = csv.DictReader(csvfile, delimiter=';')  # Specify the delimiter
        fieldnames = next(reader)  # Get the first row as fieldnames
        column_names = fieldnames['Column names'].split(';')  # Split the combined column names
        if column_name not in column_names:
            print(f"Column '{column_name}' not found in {file_path}")
            print("Available columns:", column_names)
            return

        column_names.remove(column_name)
        rows.append(column_names)

        for row in reader:
            del row[column_name]
            rows.append([row[field] for field in column_names])

    with open(file_path, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile, delimiter=';')  # Specify the delimiter
        writer.writerow(column_names)
        writer.writerows(rows)

file_path = "/content/drive/MyDrive/Copie de 12342.csv"
# Name of the column to delete
column_name = 'Perimeter'
delete_column_by_name(file_path, column_name)
```

```python
import csv
import os

# Function to read CSV files and align columns based on column names
def read_and_align_csv(file_path):
    aligned_data = []
    with open(file_path, 'r', newline='') as csvfile:
        reader = csv.DictReader(csvfile, delimiter=';')
        for row in reader:
            aligned_data.append(row)
    return aligned_data

# Define the directory where your CSV files are located
directory = '/content/drive/MyDrive/needed_rows_2'

# Get a list of all CSV files in the directory
csv_files = [os.path.join(directory, file) for file in os.listdir(directory) if file.endswith('.csv')]

# Read and align CSV files
aligned_data = []
for file in csv_files:
    aligned_data.extend(read_and_align_csv(file))

# Get a set of all keys (columns)
all_keys = set().union(*(row.keys() for row in aligned_data))

# Ensure all rows have the same keys (columns) and fill missing values with None
for row in aligned_data:
    for key in all_keys:
        row.setdefault(key, None)

# Write the merged data to a new CSV file
output_file = '/content/drive/MyDrive/complete_data.csv'
with open(output_file, 'w', newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=all_keys, delimiter=';')
    writer.writeheader()
    writer.writerows(aligned_data)
```
